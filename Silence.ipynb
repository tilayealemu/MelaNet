```python
from src.silence import split_silence
from IPython.display import display
from pydub import AudioSegment
import os


def split_audio(input_dir, file, output_dir):
    audio = AudioSegment.from_mp3(input_dir + "/" + file)
    chunks = split_silence(audio, 3000, 7000)
    count = 1
    for i in range(0, len(chunks)):
        if (i == 0):
            start = 0
        else:
            start = max(0, chunks[i-1][0])
        end = chunks[i][0]
        outfile = output_dir + "/" + file.replace(".mp3", "-"+str(count).zfill(2) + ".mp3")
        audio[start:end].export(outfile, format="mp3")
        count += 1

def split_all_audio():
    input_dir = "data/bbl/mp3"
    output_dir = "data/bbl/mp3split"
    processed = 0
    for file in os.listdir(input_dir):
        if ("mp3" in file):
            print("processing file " + str(processed) + " " + file)
            split_audio(input_dir, file, output_dir)
            processed += 1

split_all_audio()
```
